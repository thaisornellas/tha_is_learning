### Day 4

No dia 4 é feito uma analise para entender melhor como cada variável se comporta com relação a quantidade de emprestimos realizado, como por exemplo qual o vincúlo com maior emprestimos, qual o CDU mais procurado. Essa analise é feita por meio de uma função que calcula a quantidade numérica e percentual de cada variavel a ser analisada.

Além disso, foi feito também uma função extra que relaciona as variaveis, podendo analisar por exemplo quais vinculos mais procuram certo CDU ou então qual o numerico de coleção para cada biblioteca.

#### Função frequência para cada variavel

```python
def frequencia(variavel):

    # a função gera uma tabela com as quantidades numericas e percentuais dos emprestimos de acordo com a variavel passada #
    # a variavel que entra na função é a variavel do data frame #
    # variavel: str (deve ser colocada entre aspas) #

    frequencia = pd.DataFrame(dados_totais[variavel].value_counts())
    frequencia.columns = ['quantidade']
    frequencia['porcentagem'] = round(frequencia.div(frequencia.sum()) * 100)
    
    return frequencia
```

Feita a função, vamos testa-lá

```python
frequencia('tipo_vinculo_usuario')
```

Percebe-se que alunos de graduação apresentam um percentual de 78% de procura de livros em relação aos demais. Entretanto é importante considerar fatores como a quantidade de alunos de graduação com relação aos demais.

```python
frequencia('colecao')
```

Nota-se que a coleção de livros do acervo circulante são as mais procuradas.

```python
frequencia('biblioteca')
```

```python
frequencia('CDU')
```

#### Função que relaciona as variaveis

Como uma analise extra trazemos uma função que relaciona duas variáveis para observar como uma se comporta com relação a outra

```python
def relacao_variaveis(variavel_1, variavel_2, tipo):

    # função que relaciona duas variáveis #
    # variavel_1: str #
    # variavel_2: str #
    # tipo: str (porcentagem ou total) #
    
    relacao_variaveis = dados_totais.groupby([variavel_1, variavel_2]).size().unstack(fill_value=0)
    if tipo == 'porcentagem':
        relacao_variaveis = round(relacao_variaveis.div(relacao_variaveis.sum()) * 100)   
    elif tipo == 'total':
        relacao_variaveis
    return relacao_variaveis
```

Veremos em seguida alguns exemplos

```python
relacao_variaveis("biblioteca", "CDU", 'total')
```

```python
relacao_variaveis('tipo_vinculo_usuario', 'CDU', 'total')
```

```python
relacao_variaveis('colecao', 'biblioteca', 'porcentagem')
```

Para tornar a analise mais eficaz é feito uma função que gera um gráfico boxplot dado parâmetros como o dataset, x, y e o título.

#### Função geradora de gráfico boxpot

```python
def gera_bloxplot(dataset, x, y, titulo):

# função geradora de gráfico boxplot #
# recebe como parâmetros:            #
# dataset: dados                     #
# x: dados que vai na coordenada x   #
# y: dados que vai na coordenada y   #
# titulo: str - título do gráfico    #

    plt.figure(figsize=(10, 6))
    dataset.boxplot(column=y, by=x)
    plt.title(titulo)
    plt.suptitle('')
    return plt.show()
```

É feito um teste com os dados de aluno de graduação para conferir se a função está funcionando corretamente

```python
gera_bloxplot(alunos_graduacao_acervo_circulante, 'ano', 'count', 'Distribuição de emprestimos para alunos da graduação')
```
