O próximo passo é fazer a adição de uma coluna com os dados do CDU. Primeiro foi criado uma lista vazia e feito as condições para que os temas correspondentes a cada numero de localização fosse colocado de maneira correta. _Minha maior dificuldade foi utilizar os elementos de lista de forma correta_

```python
CDU = []
for i in dados_totais["localizacao"]:
    if i < 100:
        CDU.append("Generalidades. Ciência e Conhecimento")
    elif  i < 200:
        CDU.append("Filosofia e Psicologia")
    elif i < 300:
        CDU.append("Religião")
    elif i < 400:
        CDU.append("Ciências sociais")
    elif i < 500:
        CDU.append("Classe vaga")
    elif i < 600:
        CDU.append("Matemática e Ciências naturais")
    elif i < 700:
        CDU.append("Ciências aplicadas")
    elif i < 800:
        CDU.append("Belas artes")
    elif i < 900:
        CDU.append("Linguagem. Língua. Linguística")
    else:
        CDU.append("Geografia. Biografia. História")
```

É adicionado a lista *CDU* nos dados totais

```python
dados_totais["CDU"] = CDU
```

Observando os dados conferimos se a inclusão da lista foi feita de maneira correta

```python
dados_totais.head()
```

A analise seguinte é sobre a quantidade de exemplares emprestados por ano, mês e horario. Vamos primeiro separar um DataFrame com os dados de datetime englobando todos os emprestimos

```python
emprestimos_total = pd.DataFrame(dados_totais['data_emprestimo'].value_counts()).reset_index()
emprestimos_total.columns = ['data','quantidade']
emprestimos_total
```

Observando primeiro a quantidade anual, tomamos o dt.year e fazemos a soma de todos os emprestimos por ano. Para essa parte utilizamos a ferramenta do groupby (que foi dada como dica no dia 3 do desafio).

```python
emprestimos_ano = emprestimos_total.groupby(by=emprestimos_total['data'].dt.year)['quantidade'].sum()
emprestimos_ano.index.name = 'ano'
emprestimos_ano.reset_index()
```

Para a analise gráfica temos que importar a biblioteca matplotlib

```python
import matplotlib.pyplot as plt
```

### Emprestimos por ano

```python
plt.plot(emprestimos_ano)
plt.grid(True)
plt.title('Emprestimos por ano')
plt.xlabel('Ano')
plt.show()
```

Pelo gráfico é possível observar que houve um pico em 2013 e a partir desse ano começou a descer, em 2015 voltou a subir ocorrendo um pico novamente em 2017. As estratégias envolveriam saber o que aconteceu nesses anos que fez com que chegasse ao ápice e começasse a cair os emprestimos.

Dados de 2020 não estão completos visto que só temos dados da primeira metade de 2020

### Emprestimos por mês

```python
emprestimos_mes = emprestimos_total.groupby(by=emprestimos_total['data'].dt.month)['quantidade'].sum()
meses = {1: 'Jan', 2:'Fev', 3:'Mar', 4:'Abr', 5:'Mai', 6: 'Jun',7:'Jul',8:'Ago',9:'Set',10:"Out", 11:'Nov', 12:'Dez'}
emprestimos_mes.index.name = 'mes'
emprestimos_mes.index = emprestimos_mes.index.map(meses)
emprestimos_mes.reset_index()
```

```python
plt.plot(emprestimos_mes)
plt.grid(True)
plt.title('Emprestimos por mês')
plt.xlabel('Meses')
```

É possível observar que nos meses em que normalmente os alunos estão de férias (Janeiro, Julho e Dezembro) há uma baixa procura de livros na biblioteca. Além desses meses nota-se que Junho também apresenta um baixo valor de procura junto com Novembro.

Os meses com maior demanda são Março e Agosto, e com demanda consideravel em Fevereiro, Abril e Maio.

### Emprestimo por hora

```python
emprestimos_hora = emprestimos_total.groupby(by=emprestimos_total['data'].dt.hour)['quantidade'].sum()
emprestimos_hora.index.name = "hora"
emprestimos_hora = emprestimos_hora.reset_index()
emprestimos_hora
```

```python
plt.bar(emprestimos_hora['hora'], emprestimos_hora['quantidade'], color='blue', edgecolor='black')
plt.title("Emprestimos por hora")
plt.xlabel('Horário')
plt.xlim(6,23)
plt.show()
```

Nota se uma maior demanda as 10h na parte da manhã e as 16h na parte da tarde. O melhor horário para atividades que não precisam de atendimento ao usuario seriam as 8h, as 13h e as 19h.
