### *Day 2*
No dia 2 de desafio será feito limpezas de dados que não se fazem necessário, modificação da matricula de float para string e acrescentar uma nova coluna com os dados do CDU (Classificação Decimal Universal) das bibliotecas que expõe qual o tema para cada numero de localização.

Começamos analisando os dados totais

```python
dados_totais.head()
```

É feito a remoção da coluna de registro_sistema com o drop. axis = 1 quer dizer que estamos querendo remover a coluna e inplace=True é para realizarmos a operação diretamente no DataFrame original, sem ser necessário e reatribuição.

```python
dados_totais.drop("registro_sistema", axis=1, inplace=True)
```

Como nosso objetivo é transformar a coluna de *matricula ou siape* em string, por curiosidade vemos qual o tipo original dela

```python
dados_totais["matricula_ou_siape"].dtype
```

É realizado a transformação para o tipo *string*

```python
dados_totais["matricula_ou_siape"] = dados_totais["matricula_ou_siape"].astype(str)
```

Confirmando que agora a coluna é do tipo string com o dtype

```python
dados_totais["matricula_ou_siape"].dtype
```

O próximo passo é fazer a adição de uma coluna com os dados do CDU. Primeiro foi criado uma lista vazia e feito as condições para que os temas correspondentes a cada numero de localização fosse colocado de maneira correta. _Minha maior dificuldade foi utilizar os elementos de lista de forma correta_

```python
CDU = []
for i in dados_totais["localizacao"]:
    if i < 100:
        CDU.append("Generalidades. Ciência e Conhecimento")
    elif  i < 200:
        CDU.append("Filosofia e Psicologia")
    elif i < 300:
        CDU.append("Religião")
    elif i < 400:
        CDU.append("Ciências sociais")
    elif i < 500:
        CDU.append("Classe vaga")
    elif i < 600:
        CDU.append("Matemática e Ciências naturais")
    elif i < 700:
        CDU.append("Ciências aplicadas")
    elif i < 800:
        CDU.append("Belas artes")
    elif i < 900:
        CDU.append("Linguagem. Língua. Linguística")
    else:
        CDU.append("Geografia. Biografia. História")
```

É adicionado a lista *CDU* nos dados totais

```python
dados_totais["CDU"] = CDU
```
